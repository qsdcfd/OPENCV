## 자동 이진화

### 임계값 자동 결정 방법

- 영상의 히스토그램이 bimodal이고, 전경과 배경 픽셀 분포가 비슷하다

![](https://images.velog.io/images/qsdcfd/post/513fb087-4ec9-43df-bcfd-166eb8ba7926/image.png)

- 영상의 히스토그램이 bimodal이고, 전경과 배경 픽셀 분포가 다르다

![](https://images.velog.io/images/qsdcfd/post/3a5231a5-2c66-45d9-a247-3c2dab5c6f1c/image.png)

### Ostu 이진화 방법

- 입력 영상이 배경과 객체 두 개로 구성이 되어있다(Bimodal histogram)

- 임의의 임계값 T에의해 픽셀 분포 그룹의 분리

- 최적화 알고리즘

![](https://images.velog.io/images/qsdcfd/post/5ec5d598-60c9-4247-bc93-c2227b0edee5/image.png)

- Within-class variance

![](https://images.velog.io/images/qsdcfd/post/c4d0d3e2-5e4e-455b-9413-b426b5010bae/image.png)

- Betwwen-class variance 최대화

   - Within-class variance 최소화 → Between-class variance 최대화
   
   ![](https://images.velog.io/images/qsdcfd/post/19cfc035-6733-4aee-a767-ccbd53b1c2ea/image.png)

   - 모든 t 값에 대해 최적 T를 선택하게 한다 (Slow)
   
   - Recursion이용한 효율적 계산(Fast)
   
   ![](https://images.velog.io/images/qsdcfd/post/6855200a-fe0a-4221-a2e9-dd1384e024ff/image.png)
   
- Otsu 임계값

### CODE 예제

- Ostu 방법을 이용한 자동 이진화

```

src = cv2.imread('rice.png', cv2.IMREAD_GRAYSCALE)
th, dst = cv2.threshold(src, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
print("otsu's threshold:", th) # 131

```
